**Fuktionale Programmierung**

* Funktionen sind *first-class* citizens, also ganz normale Werte wie `int` und `bool`,
  * &rarr; Funktionen können als Parameter übergeben und als Rückgabewerte zurückgegeben werden,
* Funktionen sind *pure*, außer dem Rückgabewert gibt es keine Seiteneffekte,
  * &rarr; Für gleiche Parameter liefert eine Funktion immer das gleiche Ergebnis.
* Alle Variablen sind *immutable*, ihr Wert kann sich nicht ändern (außer shadowing)
  * &rarr; Es gibt keine Probleme beispielsweise bei der Parallelisierung, dafür ggf. Effizienzverlust
